# complete_given_view를 보는 이유
# - 단순 completed_rate는 '받은 전체 오퍼 중 완료 비율'이므로,
#   오퍼를 실제로 본 뒤 행동한 것인지까지는 구분하지 못한다.
# - 이번 분석에서는 퍼널 구조(received → viewed → completed)를 반영해,
#   '오퍼를 본 사람들 중 실제 완료로 이어진 비율'을 따로 확인하고자 한다.
# - 이를 통해
#   1) 오퍼를 봤을 때 구매로 잘 이어지는 고객층
#   2) 오퍼를 보지 않았는데도 완료되는 고객층
#   을 구분해 해석할 수 있다.
# - 전자는 어떤 오퍼가 효과적인지 추가 분석 대상으로,
#   후자는 프로모션 효율 재검토 대상으로 볼 수 있다.

```python
seg = (
    df
    .groupby(['gender', 'age_group', 'income_group'])
    .agg(
        n=('customer_id', 'count'),
        viewed_rate=('is_viewed', 'mean'),
        completed_rate=('is_completed', 'mean'),
        completed_without_view_rate=('completed_without_view', 'mean')
    )
    .reset_index()
)

seg['complete_given_view'] = np.where(
    seg['viewed_rate'] > 0,
    seg['completed_rate'] / seg['viewed_rate'],
    np.nan
)

seg.head()
```

```python
seg[seg['n'] >= 100].sort_values('complete_given_view', ascending=False).head(10)
```

```python
seg[seg['n'] >= 100].sort_values('completed_without_view_rate', ascending=False).head(10)
```

```python
seg['segment_share'] = seg['n'] / seg['n'].sum()
seg.head()
```
