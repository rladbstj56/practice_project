# complete_given_view를 보는 이유
 - 단순 completed_rate는 '받은 전체 오퍼 중 완료 비율'이므로,
    - 오퍼를 실제로 본 뒤 행동한 것인지까지는 구분하지 못한다.
    - 이번 분석에서는 퍼널 구조(received → viewed → completed)를 반영해,
    - '오퍼를 본 사람들 중 실제 완료로 이어진 비율'을 따로 확인하고자 한다.
- 이를 통해
   1) 오퍼를 봤을 때 구매로 잘 이어지는 고객층
   2) 오퍼를 보지 않았는데도 완료되는 고객층을 구분해 해석할 수 있다.
- 전자는 어떤 오퍼가 효과적인지 추가 분석 대상으로, 후자는 프로모션 효율 재검토 대상으로 볼 수 있다.

```python
seg = (
    df
    .groupby(['gender', 'age_group', 'income_group'])
    .agg(
        n=('customer_id', 'count'),
        viewed_rate=('is_viewed', 'mean'),
        completed_rate=('is_completed', 'mean'),
        completed_without_view_rate=('completed_without_prior_view', 'mean')
    )
    .reset_index()
)

seg['complete_given_view'] = np.where(
    seg['viewed_rate'] > 0,
    seg['completed_rate'] / seg['viewed_rate'],
    np.nan
)

seg.head()
```

```python
seg = (
    df
    .groupby(['gender', 'age_group', 'income_group'])
    .agg(
        n=('customer_id', 'count'),
        viewed=('is_viewed', 'sum'),
        completed=('is_completed', 'sum'),
        viewed_and_completed=(
            'is_completed',
            lambda x: ((x == True) & (df.loc[x.index, 'is_viewed'] == True)).sum()
        ),
        completed_without_view=('completed_without_prior_view', 'sum')
    )
    .reset_index()
)
```

```python
seg['complete_given_view'] = seg['viewed_and_completed'] / seg['viewed']
seg[seg['n'] >= 100].sort_values(
    ['complete_given_view', 'n'],
    ascending=[False, False]
).head(10)
```

#### 고객 유형 별 오퍼 분석 1 
- 고연령·고소득 세그먼트는 오퍼 노출 시 높은 전환율을 보이는 동시에, 오퍼 없이도 구매가 활발하게 발생하는 특성을 보인다. 이는 해당 집단이 가격 인센티브보다는 브랜드 충성도 및 소비 여력에 기반해 구매하는 경향이 있음을 시사한다. 
> 따라서 단순 할인 중심의 프로모션보다는 프리미엄 경험 및 리워드 중심 전략이 더 효과적일 가능성이 있다.

## 고연령 고소득 세그먼트에겐 어떤 오퍼?

```python
df['is_viewed_before_completed'] = (
    (df['is_viewed'] == True) &
    (df['is_completed'] == True) &
    (df['t_viewed'] < df['t_completed']))
```

```python
target_df = df[
    (df['age_group'].isin(['50대', '60대 이상'])) &
    (df['income_group'].isin(['7.5-10만', '10만 이상']))
].copy()

print(target_df.shape)
display(target_df.head())
```

```python
offer_seg = (
    target_df.groupby('offer_type').agg(
        n = ('customer_id', 'count'),
        viewed = ('is_viewed', 'sum'),
        completed = ('is_completed', 'sum'),
        viewed_then_completed = (
            'is_viewed_before_completed', 'sum')
    ).reset_index()
)
offer_seg
```

```python
df['is_valid_conversion'] = (
    (df['is_viewed'] == True) &
    (df['is_completed'] == True) &
    (df['t_viewed'] < df['t_completed'])
)       #보고 완료

df['is_view_only_dropout'] = (
    (df['is_viewed'] == True) &
    (df['is_completed'] == False)
)       #보고 미완료

df['is_completed_without_view'] = (
    (df['is_completed'] == True) &
    (df['is_viewed'] == False)
)       #안보고 완료
```

```python
seg = (
    df.groupby(['gender', 'age_group', 'income_group'])
    .agg(
        n=('customer_id', 'count'),
        viewed=('is_viewed', 'sum'),
        completed=('is_completed', 'sum'),
        valid_conversion=('is_valid_conversion', 'sum'),
        completed_without_view=('is_completed_without_view', 'sum'),
        view_only_dropout=('is_view_only_dropout', 'sum')
    )
    .reset_index()
)

seg['view_rate'] = seg['viewed'] / seg['n']
seg['valid_conversion_rate_per_received'] = seg['valid_conversion'] / seg['n']
seg['valid_conversion_rate_per_viewed'] = seg['valid_conversion'] / seg['viewed']
seg['completed_without_view_ratio'] = seg['completed_without_view'] / seg['completed']
seg['dropout_after_view_ratio'] = seg['view_only_dropout'] / seg['viewed']

seg.sort_values('valid_conversion', ascending=False).head(10)
```

```python
target_df = df[
    (df['age_group'].isin(['50대', '60대 이상'])) &
    (df['income_group'].isin(['7.5-10만', '10만 이상']))
].copy()

offer_seg = (
    target_df.groupby('offer_type')
    .agg(
        n=('customer_id', 'count'),
        viewed=('is_viewed', 'sum'),
        completed=('is_completed', 'sum'),
        valid_conversion=('is_valid_conversion', 'sum'),
        completed_without_view=('is_completed_without_view', 'sum')
    )
    .reset_index()
)

offer_seg['view_rate'] = offer_seg['viewed'] / offer_seg['n']
offer_seg['valid_conversion_rate_per_received'] = offer_seg['valid_conversion'] / offer_seg['n']
offer_seg['valid_conversion_rate_per_viewed'] = offer_seg['valid_conversion'] / offer_seg['viewed']
offer_seg['completed_without_view_ratio'] = offer_seg['completed_without_view'] / offer_seg['completed']
offer_seg
```

```python
# -----------------------------------
# 4) 연령대 + 성별 조합별 오퍼타입 성과 비교
# -----------------------------------

age_gender_offer_seg = (
    df.groupby(['age_group', 'gender', 'offer_type'])
    .agg(
        n=('customer_id', 'count'),
        viewed=('is_viewed', 'sum'),
        completed=('is_completed', 'sum'),
        valid_conversion=('is_valid_conversion', 'sum'),
        completed_without_view=('is_completed_without_view', 'sum'),
        view_only_dropout=('is_view_only_dropout', 'sum')
    )
    .reset_index()
)

# 비율 계산
age_gender_offer_seg['view_rate'] = age_gender_offer_seg['viewed'] / age_gender_offer_seg['n']
age_gender_offer_seg['valid_conversion_rate_per_received'] = (
    age_gender_offer_seg['valid_conversion'] / age_gender_offer_seg['n']
)
age_gender_offer_seg['valid_conversion_rate_per_viewed'] = (
    age_gender_offer_seg['valid_conversion'] / age_gender_offer_seg['viewed']
)
age_gender_offer_seg['completed_without_view_ratio'] = (
    age_gender_offer_seg['completed_without_view'] / age_gender_offer_seg['completed']
)
age_gender_offer_seg['dropout_after_view_ratio'] = (
    age_gender_offer_seg['view_only_dropout'] / age_gender_offer_seg['viewed']
)

age_gender_offer_seg
```

```python
# -----------------------------------
# 5) 연령대 + 성별 조합별 최적 오퍼 1개 뽑기
#    기준: viewed 기준 정상 전환율
# -----------------------------------

best_offer_by_age_gender = (
    age_gender_offer_seg
    .sort_values(
        ['age_group', 'gender', 'valid_conversion_rate_per_viewed'],
        ascending=[True, True, False]
    )
    .groupby(['age_group', 'gender'], as_index=False)
    .first()
)

best_offer_by_age_gender[
    [
        'age_group', 'gender', 'offer_type',
        'n', 'view_rate',
        'valid_conversion_rate_per_received',
        'valid_conversion_rate_per_viewed',
        'completed_without_view_ratio',
        'dropout_after_view_ratio'
    ]
]
```
