```python
valid_df = df[
    (df['is_viewed']==True)&
    (df['is_completed']==True)&
    (df['completed_without_prior_view']==False)
].copy()

valid_df.head(10)
```

```python
gap_seg = (
    valid_df
    .groupby(['gender', 'age_group', 'income_group'])
    .agg(
        avg_gap = ('gap_view_to_complete', 'mean'),
        median_gap = ('gap_view_to_complete', 'median'),
        n = ('customer_id', 'count')
    )
    .reset_index()
)

gap_seg.sort_values('avg_gap').head(10)
```
